```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Burn-Down-Chart.csv")
```

```python
df
```

```python
max_sprint = 4
sprint = np.array(range(1,max_sprint+1))
sum_score = df["Points"].sum()
func = lambda x : x*(sum_score/max_sprint)
y = func(sprint)
y = (sum_score-y)+24.25
```

```python
plt.plot(sprint, y, marker="o", label = "Estimeted Burndown")
plt.legend()
plt.xlabel("sprint")
plt.ylabel("score")
plt.grid(axis="x")
plt.grid(axis="y", ls='--')
plt.xticks([1,2,3,4])
plt.title("Burndown Chart")
plt.plot()
```
